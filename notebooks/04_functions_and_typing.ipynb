## 1. Why Functions Help

So far, you have written code step by step in notebook cells.

That works well when a calculation is used once.

But if we want to reuse the same logic several times, it is helpful to give that logic a name.

A function lets us do exactly that.


```python
def rainfall_mm_to_cm(rainfall_mm):
    return rainfall_mm / 10.0

rainfall_cm = rainfall_mm_to_cm(12.0)
print(rainfall_cm)
```

Here Python does two things:

1. it defines the function `rainfall_mm_to_cm`
2. it calls that function with the value `12.0`

The `return` line sends the result back out of the function.

## 2. Functions With Two Inputs

A function can use more than one input.

This is useful when a result depends on both data and a parameter.

A simple runoff-coefficient calculation is a good example.


```python
def compute_runoff_mm(rainfall_mm, runoff_coefficient):
    return rainfall_mm * runoff_coefficient

runoff_mm = compute_runoff_mm(18.0, 0.35)
print(runoff_mm)
```

This function takes two inputs:

- `rainfall_mm`
- `runoff_coefficient`

When we call the function, Python matches the input values to those parameter names and computes the result.

## 3. Functions With Conditional Logic

A function can also make decisions inside its body.

This connects directly to notebook 02.

For example, we can write a function that checks whether a measurement is valid.


```python
def is_valid_measurement(value_mm):
    if (value_mm < 0.0):
        return False
    else:
        return True

print(is_valid_measurement(4.5))
print(is_valid_measurement(-1.0))
```

Here the rule is simple:

- negative values are treated as invalid
- zero and positive values are treated as valid

This kind of rule will be useful in the final challenge.

## 4. Functions With Loops Inside

A function can also contain a loop.

This connects directly to notebook 03.

That means a function can work through a whole list and return one summary result.


```python
def total_series_mm(series_mm):
    total_mm = 0.0

    for value_mm in series_mm:
        total_mm = total_mm + value_mm

    return total_mm

print(total_series_mm([12.0, 4.5, 0.0, 7.2]))
```

The function starts a running total at `0.0`, loops through the list, and returns the final total.

So the loop is now packaged inside a reusable function.

## 5. Basic Type Hints

Type hints are extra notes that explain what kinds of values a function expects and returns.

They help humans read the code, and they can help tools check the code.

Python does not need these hints in order to run the function.


```python
def total_series_mm_typed(series_mm: list[float]) -> float:
    total_mm = 0.0

    for value_mm in series_mm:
        total_mm = total_mm + value_mm

    return total_mm

print(total_series_mm_typed([1.0, 2.5, 3.0]))
print(total_series_mm_typed.__annotations__)
```

Here the type hints say:

- `series_mm: list[float]` means the function expects a list of decimal numbers
- `-> float` means the function returns one decimal number

You do not need to memorize all of this at once. For now, read type hints as helpful labels.

## 6. Final Challenge: Clean Bad Data And Compute Summaries

Now we can bring several ideas together.

Imagine we receive a short precipitation series and a short runoff series.

Some values are bad because they are negative. In this notebook, we will use a very simple cleaning rule:

- if a value is negative, replace it with `0.0`
- otherwise keep the value as it is

Your challenge is to write several typed functions and use them together.


### Final Exercise

Use these input lists:

- `precipitation_mm = [12.0, -3.0, 0.0, 8.5, 5.0]`
- `runoff_mm = [3.0, 1.2, -0.5, 2.8, 1.0]`

Write and use these functions:

- `clean_series_mm(series_mm: list[float]) -> list[float]`
- `total_series_mm(series_mm: list[float]) -> float`
- `count_positive_days(series_mm: list[float]) -> int`
- `runoff_ratio(total_runoff_mm: float, total_precip_mm: float) -> float`
- `build_summary(total_precip_mm: float, total_runoff_mm: float, runoff_ratio_value: float) -> dict[str, float]`

Then:

1. clean both lists
2. compute total cleaned precipitation
3. compute total cleaned runoff
4. count wet days in the cleaned precipitation series
5. compute the runoff ratio
6. build a summary dictionary with the totals and the ratio

The example solution below is one clear way to do it.


```python
precipitation_mm = [12.0, -3.0, 0.0, 8.5, 5.0]
runoff_mm = [3.0, 1.2, -0.5, 2.8, 1.0]


def clean_series_mm(series_mm: list[float]) -> list[float]:
    """Return a cleaned series where negative values are replaced with 0.0."""
    cleaned_series_mm = []

    for value_mm in series_mm:
        if (value_mm < 0.0):
            cleaned_series_mm.append(0.0)
        else:
            cleaned_series_mm.append(value_mm)

    return cleaned_series_mm



def total_series_mm(series_mm: list[float]) -> float:
    """Return the total of all values in a series."""
    total_mm = 0.0

    for value_mm in series_mm:
        total_mm = total_mm + value_mm

    return total_mm



def count_positive_days(series_mm: list[float]) -> int:
    """Return how many values in a series are greater than 0.0."""
    count = 0

    for value_mm in series_mm:
        if (value_mm > 0.0):
            count = count + 1

    return count



def runoff_ratio(total_runoff_mm: float, total_precip_mm: float) -> float:
    """Return runoff divided by precipitation."""
    return total_runoff_mm / total_precip_mm



def build_summary(total_precip_mm: float, total_runoff_mm: float, runoff_ratio_value: float) -> dict[str, float]:
    """Return a summary dictionary with precipitation, runoff, and runoff ratio."""
    return {
        "total_precip_mm": total_precip_mm,
        "total_runoff_mm": total_runoff_mm,
        "runoff_ratio": runoff_ratio_value,
    }


cleaned_precipitation_mm = clean_series_mm(precipitation_mm)
cleaned_runoff_mm = clean_series_mm(runoff_mm)

total_precip_mm = total_series_mm(cleaned_precipitation_mm)
total_runoff_mm = total_series_mm(cleaned_runoff_mm)
wet_day_count = count_positive_days(cleaned_precipitation_mm)
runoff_ratio_value = runoff_ratio(total_runoff_mm, total_precip_mm)
summary = build_summary(total_precip_mm, total_runoff_mm, runoff_ratio_value)

print(cleaned_precipitation_mm)
print(cleaned_runoff_mm)
print(total_precip_mm)
print(total_runoff_mm)
print(wet_day_count)
print(runoff_ratio_value)
print(summary)
```

```python
# You do not need to understand this check code in detail.
# What matters is that the cell runs without an error.
assert clean_series_mm([2.0, -1.0, 0.5]) == [2.0, 0.0, 0.5], "clean_series_mm should replace negative values with 0.0."
assert cleaned_precipitation_mm == [12.0, 0.0, 0.0, 8.5, 5.0], "Check your cleaned precipitation series."
assert cleaned_runoff_mm == [3.0, 1.2, 0.0, 2.8, 1.0], "Check your cleaned runoff series."
assert total_precip_mm == 25.5, "The cleaned precipitation total should be 25.5."
assert total_runoff_mm == 8.0, "The cleaned runoff total should be 8.0."
assert wet_day_count == 3, "There should be 3 wet days in the cleaned precipitation series."
assert abs(runoff_ratio_value - (8.0 / 25.5)) < 1e-9, "Check your runoff ratio calculation."
assert summary == {"total_precip_mm": 25.5, "total_runoff_mm": 8.0, "runoff_ratio": 8.0 / 25.5}, "Check the summary dictionary values."
print("Final exercise check passed.")
```
