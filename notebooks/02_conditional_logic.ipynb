## 1. Comparisons Become Booleans

A comparison asks a question.

For example:

- is rainfall greater than zero?
- is temperature below freezing?
- is snow depth equal to zero?

Python answers these questions with a boolean: either `True` or `False`.


```python
rainfall_mm = 6.2
temperature_c = -1.0

is_wet_day = (rainfall_mm > 0.0)
is_freezing = (temperature_c < 0.0)

print(is_wet_day)
print(is_freezing)
print(type(is_wet_day))
```

In this output:

- `True` means the comparison was correct
- `False` would mean the comparison was not correct
- `<class 'bool'>` means Python created a boolean value

So `rainfall_mm > 0.0` is not just text. Python evaluates it and turns it into either `True` or `False`.


Python has several comparison operators:

- `>` means greater than
- `<` means less than
- `>=` means greater than or equal to
- `<=` means less than or equal to
- `==` means equal to


```python
rainfall_mm = 0.0
temperature_c = 0.0
discharge_m3s = 12.5
expected_discharge_m3s = 12.5

print(rainfall_mm > 0.0)
print(temperature_c < 0.0)
print(temperature_c >= 0.0)
print(rainfall_mm <= 1.0)
print(discharge_m3s == expected_discharge_m3s)
```

Be careful with these two symbols:

- `=` means assign a value to a variable
- `==` means compare two values and ask whether they are equal

Example:

- `temperature_c = 0.0` stores `0.0` in the variable `temperature_c`
- `temperature_c == 0.0` asks whether `temperature_c` is equal to `0.0`

## 2. Using `if`

An `if` statement lets Python run code only when a condition is `True`.

Python reads this pattern like this:

1. check the condition after `if`
2. if the answer is `True`, run the indented code block
3. if the answer is `False`, skip that block

The colon `:` starts the block, and the indentation shows which lines belong to that block.


```python
rainfall_mm = 8.0

if (rainfall_mm > 0.0):
    print("This is a wet day.")

print("The check is finished.")
```

Here, `rainfall_mm > 0.0` is `True`, so Python prints `This is a wet day.`

Then Python continues and prints `The check is finished.`

If the condition had been `False`, only the indented line would have been skipped.

## 3. Using `if` And `else`

Sometimes we want one message when a condition is `True` and a different message when it is `False`.

That is what `else` is for.


```python
temperature_c = -1.5

if temperature_c <= 0.0:
    precipitation_type = "snow"
else:
    precipitation_type = "rain"

print(precipitation_type)
```

This example uses a very simple threshold rule:

- at or below `0.0`, we label the precipitation as `snow`
- above `0.0`, we label it as `rain`

Because `temperature_c` is `-1.5`, the `if` branch runs and `precipitation_type` becomes `snow`.

## 4. Using `if`, `elif`, And `else`

Sometimes we have more than two possibilities.

In that case, `elif` lets us test another condition if the earlier one was not true.

Python checks the branches from top to bottom and runs only the first branch that matches.


```python
discharge_m3s = 28.0

if (discharge_m3s < 10.0):
    discharge_label = "low"
elif (discharge_m3s < 30.0):
    discharge_label = "medium"
else:
    discharge_label = "high"

print(discharge_label)
```

Here Python checks the branches in order:

- `discharge_m3s < 10.0` is `False`
- `discharge_m3s < 30.0` is `True`
- so Python stops there and sets `discharge_label` to `medium`

The `else` branch is only used when the earlier conditions were not true.
