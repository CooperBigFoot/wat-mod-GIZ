## 1. Comparisons Become Booleans

A comparison asks a question.

For example:

- is rainfall greater than zero?
- is temperature below freezing?
- is snow depth equal to zero?

Python answers these questions with a boolean: either `True` or `False`.


## 1. Сравнения превращаются в булевы значения

Сравнение задает вопрос.

Например:

- больше ли количество осадков нуля?
- ниже ли температура точки замерзания?
- равна ли глубина снега нулю?

Python отвечает на такие вопросы булевым значением: либо `True`, либо `False`.


```python
rainfall_mm = 6.2
temperature_c = -1.0

is_wet_day = (rainfall_mm > 0.0)
is_freezing = (temperature_c < 0.0)

print(is_wet_day)
print(is_freezing)
print(type(is_wet_day))
```

In this output:

- `True` means the comparison was correct
- `False` would mean the comparison was not correct
- `<class 'bool'>` means Python created a boolean value

So `rainfall_mm > 0.0` is not just text. Python evaluates it and turns it into either `True` or `False`.


В этом выводе:

- `True` означает, что сравнение оказалось верным
- `False` означало бы, что сравнение оказалось неверным
- `<class 'bool'>` означает, что Python создал булево значение

Поэтому `rainfall_mm > 0.0` — это не просто текст. Python вычисляет это выражение и превращает его либо в `True`, либо в `False`.


Python has several comparison operators:

- `>` means greater than
- `<` means less than
- `>=` means greater than or equal to
- `<=` means less than or equal to
- `==` means equal to


В Python есть несколько операторов сравнения:

- `>` означает больше чем
- `<` означает меньше чем
- `>=` означает больше или равно
- `<=` означает меньше или равно
- `==` означает равно


```python
rainfall_mm = 0.0
temperature_c = 0.0
discharge_m3s = 12.5
expected_discharge_m3s = 12.5

print(rainfall_mm > 0.0)
print(temperature_c < 0.0)
print(temperature_c >= 0.0)
print(rainfall_mm <= 1.0)
print(discharge_m3s == expected_discharge_m3s)
```

Be careful with these two symbols:

- `=` means assign a value to a variable
- `==` means compare two values and ask whether they are equal

Example:

- `temperature_c = 0.0` stores `0.0` in the variable `temperature_c`
- `temperature_c == 0.0` asks whether `temperature_c` is equal to `0.0`


Будьте внимательны с этими двумя символами:

- `=` означает присвоить значение переменной
- `==` означает сравнить два значения и спросить, равны ли они

Пример:

- `temperature_c = 0.0` сохраняет `0.0` в переменной `temperature_c`
- `temperature_c == 0.0` спрашивает, равно ли `temperature_c` значению `0.0`

## 2. Using `if`

An `if` statement lets Python run code only when a condition is `True`.

Python reads this pattern like this:

1. check the condition after `if`
2. if the answer is `True`, run the indented code block
3. if the answer is `False`, skip that block

The colon `:` starts the block, and the indentation shows which lines belong to that block.


## 2. Использование `if`

Инструкция `if` позволяет Python выполнять код только тогда, когда условие равно `True`.

Python читает этот шаблон так:

1. проверить условие после `if`
2. если ответ `True`, выполнить блок кода с отступом
3. если ответ `False`, пропустить этот блок

Двоеточие `:` начинает блок, а отступ показывает, какие строки относятся к этому блоку.


```python
rainfall_mm = 8.0

if (rainfall_mm > 0.0):
    print("This is a wet day.")

print("The check is finished.")
```

Here, `rainfall_mm > 0.0` is `True`, so Python prints `This is a wet day.`

Then Python continues and prints `The check is finished.`

If the condition had been `False`, only the indented line would have been skipped.


Здесь `rainfall_mm > 0.0` равно `True`, поэтому Python печатает `This is a wet day.`

Затем Python продолжает выполнение и печатает `The check is finished.`

Если бы условие было `False`, была бы пропущена только строка с отступом.

## 3. Using `if` And `else`

Sometimes we want one message when a condition is `True` and a different message when it is `False`.

That is what `else` is for.


## 3. Использование `if` и `else`

Иногда нам нужно одно сообщение, когда условие равно `True`, и другое сообщение, когда оно равно `False`.

Для этого используется `else`.


```python
temperature_c = -1.5

if temperature_c <= 0.0:
    precipitation_type = "snow"
else:
    precipitation_type = "rain"

print(precipitation_type)
```

This example uses a very simple threshold rule:

- at or below `0.0`, we label the precipitation as `snow`
- above `0.0`, we label it as `rain`

Because `temperature_c` is `-1.5`, the `if` branch runs and `precipitation_type` becomes `snow`.


В этом примере используется очень простое пороговое правило:

- при `0.0` или ниже мы обозначаем осадки как `snow`
- выше `0.0` мы обозначаем их как `rain`

Поскольку `temperature_c` равно `-1.5`, выполняется ветвь `if`, и `precipitation_type` становится `snow`.

## 4. Using `if`, `elif`, And `else`

Sometimes we have more than two possibilities.

In that case, `elif` lets us test another condition if the earlier one was not true.

Python checks the branches from top to bottom and runs only the first branch that matches.


## 4. Использование `if`, `elif` и `else`

Иногда у нас больше двух вариантов.

В этом случае `elif` позволяет проверить еще одно условие, если предыдущее не было истинным.

Python проверяет ветви сверху вниз и выполняет только первую подходящую ветвь.


```python
discharge_m3s = 28.0

if (discharge_m3s < 10.0):
    discharge_label = "low"
elif (discharge_m3s < 30.0):
    discharge_label = "medium"
else:
    discharge_label = "high"

print(discharge_label)
```

Here Python checks the branches in order:

- `discharge_m3s < 10.0` is `False`
- `discharge_m3s < 30.0` is `True`
- so Python stops there and sets `discharge_label` to `medium`

The `else` branch is only used when the earlier conditions were not true.


Здесь Python проверяет ветви по порядку:

- `discharge_m3s < 10.0` равно `False`
- `discharge_m3s < 30.0` равно `True`
- поэтому Python останавливается здесь и присваивает `discharge_label` значение `medium`

Ветвь `else` используется только тогда, когда предыдущие условия не были истинными.
