## 2. Numbers: `int` and `float`

In Python, numbers usually appear in two basic forms:

- `int` for whole numbers like `2`
- `float` for decimal numbers like `18.7`

Hydrological measurements are often floats because they are not usually whole numbers.

We can also do arithmetic with numbers:

- `+` means add
- `-` means subtract
- `*` means multiply
- `/` means divide

```python
n_stations = 2
daily_rainfall_mm = 18.7
next_day_rainfall_mm = 6.3
total_rainfall_mm = daily_rainfall_mm + next_day_rainfall_mm

print(type(n_stations))
print(type(daily_rainfall_mm))
print(total_rainfall_mm)
```

Why are we printing `type(...)`?

`type(variable)` shows us what kind of object Python thinks that variable is.

So in the output:

- `<class 'int'>` means Python sees `n_stations` as a whole number
- `<class 'float'>` means Python sees `daily_rainfall_mm` as a decimal number
- `25.0` is the result of the addition

Do not worry about the exact text `<class ...>` yet. For now, just read it as
"Python says this is an `int`" or "Python says this is a `float`".

## 3. Strings and Booleans

A `str` stores text. A `bool` stores either `True` or `False`.

Strings are useful for names and labels. Booleans are useful when a question has a yes/no answer.

A comparison like `rainfall_mm > 0.0` asks a question: is rainfall greater than zero? The answer to that question is a boolean.

```python
station_code = "ALA001"
rainfall_mm = 9.0
is_wet_day = (rainfall_mm > 0.0)  # This comparison becomes either True or False.

print(station_code)
print(is_wet_day)
print(type(station_code))
print(type(is_wet_day))
```

It is useful to separate the two symbols mentally:

- `=` means "store the result in this variable"
- `>` means "compare the left value to the right value"

So this line:

`is_wet_day = (rainfall_mm > 0.0)`

means:

1. compare `rainfall_mm` to `0.0`
2. get either `True` or `False`
3. store that answer in `is_wet_day`

## 4. Lists

A list stores several values in order.

Lists are written with square brackets: `[]`.

Lists are useful for short timeseries such as rainfall over a few days.

You can:

- access one value with an index like `[0]`
- access the last value with `[-1]`
- access part of the list with slicing like `[1:3]`

Python starts counting at `0`, not at `1`.

```python
rainfall_series_mm = [12.0, 4.5, 0.0, 7.2]

print(rainfall_series_mm)
print(rainfall_series_mm[0])   # First value
print(rainfall_series_mm[-1])  # Last value
print(rainfall_series_mm[1:3]) # A slice from index 1 up to, but not including, index 3
print(len(rainfall_series_mm)) # Number of values in the list
print(sum(rainfall_series_mm)) # Sum of all values in the list
```

The slice `rainfall_series_mm[1:3]` means:

- start at index `1`
- stop before index `3`

So it returns the values at indices `1` and `2`.

You do not need to memorize every detail yet. The important idea is that a list lets us store multiple values in order and access the parts we need.

## 5. Dictionaries

A dictionary stores named values using keys.

A dictionary is written with curly braces: `{}`.

Inside the braces, we usually write entries in a `key: value` form, for example:

- `"name": "Ala-Archa"`
- `"area_km2": 233.0`

Dictionaries are useful when several values belong together, such as basin metadata.

A dictionary can also store more complex things, including a list.

Lists are good when order matters. Dictionaries are good when names matter.

```python
basin = {
    "name": "Ala-Archa",
    "country": "Kyrgyzstan",
    "area_km2": 233.0,
    "has_glacier": True,
    "pet_series_mm": [2.1, 2.4, 1.9],
}

print(basin)
print(basin["name"])
print(basin["area_km2"])
print(basin["pet_series_mm"])
print(type(basin))
```

Notice that this dictionary stores different kinds of values under different keys:

- text under `"name"`
- a number under `"area_km2"`
- a boolean under `"has_glacier"`
- a list under `"pet_series_mm"`

So dictionaries are very flexible: they can store simple values, but also containers like lists.

## 6. Checking Types With `type(...)`

Python objects have types. We can inspect them with `type(...)`.

This is useful when you are learning, because it helps you see what kind of object you are working with.

As we hinted at already, when you run `type(...)`, Python prints something like `<class 'int'>`.
Do not worry about the exact format yet. The important part is the word at the end: `int`, `float`, `str`, `bool`, `list`, or `dict`.

```python
print(type(4))
print(type(4.0))
print(type("runoff"))
print(type(True))
print(type([1.0, 2.0, 3.0]))
print(type({"x": 1}))
```

You can read the output like this:

- `<class 'int'>` means "this is an integer"
- `<class 'float'>` means "this is a decimal number"
- `<class 'str'>` means "this is text"
- `<class 'bool'>` means "this is a True/False value"
- `<class 'list'>` means "this is a list"
- `<class 'dict'>` means "this is a dictionary"

## 7. Notebook State And Reassignment

A notebook remembers values from earlier cells.

If you change a variable and run the next cell again, the result can change.
This is powerful, but it also means execution order matters.

```python
runoff_coefficient = 0.35
rainfall_mm = 20.0
runoff_mm = runoff_coefficient * rainfall_mm
print(runoff_mm)

runoff_coefficient = 0.50
runoff_mm = runoff_coefficient * rainfall_mm
print(runoff_mm)
```
