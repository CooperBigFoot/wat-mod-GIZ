## 2. Numbers: `int` and `float`

In Python, numbers usually appear in two basic forms:

- `int` for whole numbers like `2`
- `float` for decimal numbers like `18.7`

Hydrological measurements are often floats because they are not usually whole numbers.

We can also do arithmetic with numbers:

- `+` means add
- `-` means subtract
- `*` means multiply
- `/` means divide

## 2. Числа: `int` и `float`

В Python числа обычно встречаются в двух основных формах:

- `int` для целых чисел, например `2`
- `float` для десятичных чисел, например `18.7`

Гидрологические измерения часто являются числами типа `float`, потому что обычно они не являются целыми числами.

С числами также можно выполнять арифметические операции:

- `+` означает сложение
- `-` означает вычитание
- `*` означает умножение
- `/` означает деление

```python
n_stations = 2
daily_rainfall_mm = 18.7
next_day_rainfall_mm = 6.3
total_rainfall_mm = daily_rainfall_mm + next_day_rainfall_mm

print(type(n_stations))
print(type(daily_rainfall_mm))
print(total_rainfall_mm)
```

Why are we printing `type(...)`?

`type(variable)` shows us what kind of object Python thinks that variable is.

So in the output:

- `<class 'int'>` means Python sees `n_stations` as a whole number
- `<class 'float'>` means Python sees `daily_rainfall_mm` as a decimal number
- `25.0` is the result of the addition

Do not worry about the exact text `<class ...>` yet. For now, just read it as
"Python says this is an `int`" or "Python says this is a `float`".

Зачем мы печатаем `type(...)`?

`type(variable)` показывает, каким типом объекта Python считает эту переменную.

Поэтому в выводе:

- `<class 'int'>` означает, что Python видит `n_stations` как целое число
- `<class 'float'>` означает, что Python видит `daily_rainfall_mm` как десятичное число
- `25.0` — это результат сложения

Пока не беспокойтесь о точном виде текста `<class ...>`. Сейчас просто читайте это как «Python говорит, что это `int`» или «Python говорит, что это `float`».

## 3. Strings and Booleans

A `str` stores text. A `bool` stores either `True` or `False`.

Strings are useful for names and labels. Booleans are useful when a question has a yes/no answer.

A comparison like `rainfall_mm > 0.0` asks a question: is rainfall greater than zero? The answer to that question is a boolean.

## 3. Строки и булевы значения

`str` хранит текст. `bool` хранит либо `True`, либо `False`.

Строки полезны для названий и меток. Булевы значения полезны, когда вопрос имеет ответ да/нет.

Сравнение вроде `rainfall_mm > 0.0` задает вопрос: больше ли количество осадков нуля? Ответ на этот вопрос является булевым значением.

```python
station_code = "ALA001"
rainfall_mm = 9.0
is_wet_day = (rainfall_mm > 0.0)  # This comparison becomes either True or False.

print(station_code)
print(is_wet_day)
print(type(station_code))
print(type(is_wet_day))
```

It is useful to separate the two symbols mentally:

- `=` means "store the result in this variable"
- `>` means "compare the left value to the right value"

So this line:

`is_wet_day = (rainfall_mm > 0.0)`

means:

1. compare `rainfall_mm` to `0.0`
2. get either `True` or `False`
3. store that answer in `is_wet_day`

Полезно мысленно разделять эти два символа:

- `=` означает «сохранить результат в этой переменной»
- `>` означает «сравнить значение слева со значением справа»

Поэтому эта строка:

`is_wet_day = (rainfall_mm > 0.0)`

означает:

1. сравнить `rainfall_mm` с `0.0`
2. получить либо `True`, либо `False`
3. сохранить этот ответ в `is_wet_day`

## 4. Lists

A list stores several values in order.

Lists are written with square brackets: `[]`.

Lists are useful for short timeseries such as rainfall over a few days.

You can:

- access one value with an index like `[0]`
- access the last value with `[-1]`
- access part of the list with slicing like `[1:3]`

Python starts counting at `0`, not at `1`.

## 4. Списки

Список хранит несколько значений в определенном порядке.

Списки записываются в квадратных скобках: `[]`.

Списки полезны для коротких временных рядов, например осадков за несколько дней.

Вы можете:

- получить одно значение по индексу, например `[0]`
- получить последнее значение с помощью `[-1]`
- получить часть списка с помощью среза, например `[1:3]`

Python начинает счет с `0`, а не с `1`.

```python
rainfall_series_mm = [12.0, 4.5, 0.0, 7.2]

print(rainfall_series_mm)
print(rainfall_series_mm[0])   # First value
print(rainfall_series_mm[-1])  # Last value
print(rainfall_series_mm[1:3]) # A slice from index 1 up to, but not including, index 3
print(len(rainfall_series_mm)) # Number of values in the list
print(sum(rainfall_series_mm)) # Sum of all values in the list
```

The slice `rainfall_series_mm[1:3]` means:

- start at index `1`
- stop before index `3`

So it returns the values at indices `1` and `2`.

You do not need to memorize every detail yet. The important idea is that a list lets us store multiple values in order and access the parts we need.

Срез `rainfall_series_mm[1:3]` означает:

- начать с индекса `1`
- остановиться перед индексом `3`

То есть он возвращает значения с индексами `1` и `2`.

Пока не нужно запоминать все детали. Главная идея в том, что список позволяет хранить несколько значений в порядке и получать нужные части.

## 5. Dictionaries

A dictionary stores named values using keys.

A dictionary is written with curly braces: `{}`.

Inside the braces, we usually write entries in a `key: value` form, for example:

- `"name": "Ala-Archa"`
- `"area_km2": 233.0`

Dictionaries are useful when several values belong together, such as basin metadata.

A dictionary can also store more complex things, including a list.

Lists are good when order matters. Dictionaries are good when names matter.

## 5. Словари

Словарь хранит именованные значения с помощью ключей.

Словарь записывается в фигурных скобках: `{}`.

Внутри скобок мы обычно записываем элементы в форме `key: value`, например:

- `"name": "Ala-Archa"`
- `"area_km2": 233.0`

Словари полезны, когда несколько значений относятся к одному объекту, например к метаданным бассейна.

Словарь также может хранить более сложные объекты, включая список.

Списки удобны, когда важен порядок. Словари удобны, когда важны имена.

```python
basin = {
    "name": "Ala-Archa",
    "country": "Kyrgyzstan",
    "area_km2": 233.0,
    "has_glacier": True,
    "pet_series_mm": [2.1, 2.4, 1.9],
}

print(basin)
print(basin["name"])
print(basin["area_km2"])
print(basin["pet_series_mm"])
print(type(basin))
```

Notice that this dictionary stores different kinds of values under different keys:

- text under `"name"`
- a number under `"area_km2"`
- a boolean under `"has_glacier"`
- a list under `"pet_series_mm"`

So dictionaries are very flexible: they can store simple values, but also containers like lists.

Обратите внимание, что этот словарь хранит разные виды значений под разными ключами:

- текст под ключом `"name"`
- число под ключом `"area_km2"`
- булево значение под ключом `"has_glacier"`
- список под ключом `"pet_series_mm"`

Поэтому словари очень гибкие: они могут хранить простые значения, а также контейнеры, например списки.

## 6. Checking Types With `type(...)`

Python objects have types. We can inspect them with `type(...)`.

This is useful when you are learning, because it helps you see what kind of object you are working with.

As we hinted at already, when you run `type(...)`, Python prints something like `<class 'int'>`.
Do not worry about the exact format yet. The important part is the word at the end: `int`, `float`, `str`, `bool`, `list`, or `dict`.

## 6. Проверка типов с помощью `type(...)`

Объекты Python имеют типы. Мы можем посмотреть их с помощью `type(...)`.

Это полезно во время обучения, потому что помогает увидеть, с каким типом объекта вы работаете.

Как мы уже упоминали, при запуске `type(...)` Python выводит что-то вроде `<class 'int'>`.
Пока не беспокойтесь о точном формате. Важная часть — слово в конце: `int`, `float`, `str`, `bool`, `list` или `dict`.

```python
print(type(4))
print(type(4.0))
print(type("runoff"))
print(type(True))
print(type([1.0, 2.0, 3.0]))
print(type({"x": 1}))
```

You can read the output like this:

- `<class 'int'>` means "this is an integer"
- `<class 'float'>` means "this is a decimal number"
- `<class 'str'>` means "this is text"
- `<class 'bool'>` means "this is a True/False value"
- `<class 'list'>` means "this is a list"
- `<class 'dict'>` means "this is a dictionary"


Вы можете читать вывод так:

- `<class 'int'>` означает «это целое число»
- `<class 'float'>` означает «это десятичное число»
- `<class 'str'>` означает «это текст»
- `<class 'bool'>` означает «это значение True/False»
- `<class 'list'>` означает «это список»
- `<class 'dict'>` означает «это словарь»

## 7. Notebook State And Reassignment

A notebook remembers values from earlier cells.

If you change a variable and run the next cell again, the result can change.
This is powerful, but it also means execution order matters.

## 7. Состояние ноутбука и переназначение переменных

Ноутбук запоминает значения из предыдущих ячеек.

Если вы измените переменную и снова запустите следующую ячейку, результат может измениться.
Это мощная возможность, но она также означает, что порядок выполнения ячеек имеет значение.

```python
runoff_coefficient = 0.35
rainfall_mm = 20.0
runoff_mm = runoff_coefficient * rainfall_mm
print(runoff_mm)

runoff_coefficient = 0.50
runoff_mm = runoff_coefficient * rainfall_mm
print(runoff_mm)
```
