## 1. `for` Loops Over Lists

A `for` loop lets Python do the same kind of action for each value in a list.

This is useful in hydrology because short time series are often stored as lists.

For example, we might want to look at rainfall for each day, one day at a time.


## 1. Циклы `for` по спискам

Цикл `for` позволяет Python выполнить действие одного и того же типа для каждого значения в списке.

Это полезно в гидрологии, потому что короткие временные ряды часто хранятся как списки.

Например, мы можем захотеть посмотреть количество осадков за каждый день, по одному дню за раз.


```python
rainfall_series_mm = [12.0, 4.5, 0.0, 7.2]

for rainfall_mm in rainfall_series_mm:
    print(rainfall_mm)
```

Python takes one value from the list, stores it in `rainfall_mm`, and runs the indented line.

Then it repeats that process for the next value.

So the loop variable `rainfall_mm` changes from one pass to the next.


Python берет одно значение из списка, сохраняет его в `rainfall_mm` и выполняет строку с отступом.

Затем он повторяет этот процесс для следующего значения.

Поэтому переменная цикла `rainfall_mm` меняется от одного прохода к следующему.

## 2. Running Totals Inside A Loop

A loop becomes more useful when we update a variable each time the loop runs.

A common pattern is a running total:

1. start with `0.0`
2. add one value each time through the loop
3. read the final total after the loop finishes


## 2. Накопительные суммы внутри цикла

Цикл становится полезнее, когда мы обновляем переменную при каждом его выполнении.

Распространенный шаблон — накопительная сумма:

1. начать с `0.0`
2. добавлять одно значение при каждом проходе цикла
3. прочитать итоговую сумму после завершения цикла


```python
rainfall_series_mm = [12.0, 4.5, 0.0, 7.2]
total_rainfall_mm = 0.0

for rainfall_mm in rainfall_series_mm:
    total_rainfall_mm = total_rainfall_mm + rainfall_mm

print(total_rainfall_mm)
```

Here `total_rainfall_mm` keeps growing as Python moves through the list.

After the loop ends, `total_rainfall_mm` holds the total rainfall for the full series.


Здесь `total_rainfall_mm` постепенно растет, пока Python проходит по списку.

После завершения цикла `total_rainfall_mm` содержит суммарное количество осадков за весь ряд.

## 3. Counting With `if` Inside A Loop

We can combine a loop with conditional logic from notebook 02.

This lets us count how many values match a condition.

For example, we can count wet days by checking whether rainfall is greater than zero.


## 3. Подсчет с `if` внутри цикла

Мы можем объединить цикл с условной логикой из ноутбука 02.

Это позволяет считать, сколько значений удовлетворяют условию.

Например, мы можем посчитать влажные дни, проверяя, больше ли количество осадков нуля.


```python
rainfall_series_mm = [12.0, 4.5, 0.0, 7.2, 0.0]
wet_day_count = 0

for rainfall_mm in rainfall_series_mm:
    if (rainfall_mm > 0.0):
        wet_day_count = wet_day_count + 1

print(wet_day_count)
```

The counter starts at `0`.

Each time Python finds a wet day, it adds `1`.

Days with `0.0` rainfall do not change the counter.


Счетчик начинается с `0`.

Каждый раз, когда Python находит влажный день, он добавляет `1`.

Дни с количеством осадков `0.0` не изменяют счетчик.

## 4. Building A New List With `append()`

Sometimes we do not want one final number.

Instead, we want to build a new list from an old list.

A common beginner pattern is:

1. start with an empty list: `[]`
2. loop through the old list
3. use `append()` to add one new value at a time


## 4. Построение нового списка с помощью `append()`

Иногда нам нужен не один итоговый числовой результат.

Вместо этого мы хотим построить новый список из старого списка.

Распространенный шаблон для начинающих:

1. начать с пустого списка: `[]`
2. пройти циклом по старому списку
3. использовать `append()`, чтобы добавлять по одному новому значению


```python
temperature_series_c = [-2.0, 1.5, 4.0]
phase_labels = []

for temperature_c in temperature_series_c:
    if (temperature_c <= 0.0):
        phase_labels.append("snow")
    else:
        phase_labels.append("rain")

print(phase_labels)
```

This loop creates one label for each temperature.

Because the new values are added in order, the new list lines up with the original temperature list.


Этот цикл создает одну метку для каждой температуры.

Поскольку новые значения добавляются по порядку, новый список соответствует исходному списку температур.

## 5. List Comprehensions

A list comprehension is a shorter way to build a new list.

It does the same kind of job as the loop-and-`append()` pattern you just used.

For beginners, it helps to read the long version first and the short version second.


## 5. Списковые включения

Списковое включение — это более короткий способ построить новый список.

Оно выполняет работу того же типа, что и шаблон с циклом и `append()`, который вы только что использовали.

Начинающим полезно сначала прочитать длинную версию, а затем короткую.


```python
rainfall_series_mm = [12.0, 4.5, 0.0, 7.2]

rainfall_series_cm_loop = []
for rainfall_mm in rainfall_series_mm:
    rainfall_series_cm_loop.append(rainfall_mm / 10.0)

rainfall_series_cm_comp = [rainfall_mm / 10.0 for rainfall_mm in rainfall_series_mm]

print(rainfall_series_cm_loop)
print(rainfall_series_cm_comp)
```

Both results are the same.

The list comprehension is shorter, but the idea is not new:

- take one value from the old list
- transform it
- store the transformed values in a new list


Оба результата одинаковые.

Списковое включение короче, но идея не новая:

- взять одно значение из старого списка
- преобразовать его
- сохранить преобразованные значения в новом списке

## 6. A Very Simple Dictionary Comprehension

A dictionary comprehension is similar to a list comprehension, but it builds a dictionary instead of a list.

We will keep this first example very small.


## 6. Очень простое словарное включение

Словарное включение похоже на списковое включение, но строит словарь вместо списка.

Этот первый пример мы сделаем очень маленьким.


```python
day_names = ["day_1", "day_2", "day_3"]

initial_rainfall_mm_loop = {}
for day_name in day_names:
    initial_rainfall_mm_loop[day_name] = 0.0

initial_rainfall_mm_comp = {day_name: 0.0 for day_name in day_names}

print(initial_rainfall_mm_loop)
print(initial_rainfall_mm_comp)
```

Again, both results are the same.

The main new idea is the `key: value` part inside the braces.

For now, read this as: "for each day name, make a dictionary entry with rainfall `0.0`."


И снова оба результата одинаковые.

Главная новая идея — часть `key: value` внутри фигурных скобок.

Пока читайте это так: «для каждого названия дня создать запись словаря с количеством осадков `0.0`».
